```python
# Import Selenium and its sub libraries
import selenium 
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import Select
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import re
```

```python
url="https://issuu.com/"
```

```python
def extract_page_data(driver):
    contents=driver.find_elements("class name","PublicationCard__publication-card__dZd0M__0-0-2365")
    print(len(contents))

    #Create loop to itterate on the list
    for con in range(len(contents)):
    
        title = contents[con].find_element("class name","PublicationCard__publication-card__card-title__jufAN__0-0-2365").text
        
        author = contents[con].find_element("class name","PublicationCard__publication-card__author-link__-bT0k__0-0-2365").text
        
        date = contents[con].find_element("class name","PublicationCard__publication-card__meta__aO2ik__0-0-2365").text
        
        link = contents[con].find_element("class name","PublicationCard__publication-card__card-link__hUKEG__0-0-2365").get_attribute('href')
        
        cover = contents[con].find_element("class name","PublicationCard__publication-card__cover-img__5ibKR__0-0-2365").get_attribute('src')        
        #page= contents[con].find_element("class name","PublicationCard__publication-card__meta__aO2ik__0-0-2365").text
        
        print(str(con)+"\n"+title+"\n"+author+"\n"+date+"\n"+link+"\n"+cover+"\n")
        
        title_list.append(title)
        author_list.append(author)
        date_list.append(date)
        link_list.append(link)
        cover_list.append(cover)
        
        sleep(2)
```

```python
driver = webdriver.Chrome()
driver.get(url)

cover_list=[]
title_list=[]
author_list=[]
link_list=[]
date_list=[]

search_query = driver.find_element("id","search")

search_query.send_keys("ucl")
search_query.send_keys(Keys.RETURN)
sleep(2)



extract_page_data(driver)

#for i in range(0,3):
    #extract_page_data(driver)
    #driver.find_element("xpath", "//*[@id="app"]/div/div[2]/div[2]/button").click()
    #i+=1
```

```python
import csv 
import requests
data=zip(title_list,author_list,date_list,link_list,cover_list)

with open('issuu.csv', 'w',newline='') as csvfile: 
    writer = csv.writer(csvfile) 
    for row in data:
        writer.writerow(row)

i=0
with open("issuu.csv",'r',encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)    
    for row in reader:
        response=requests.get(row[4])
        imagename="image"+str(i)+".jpg"
        with open(imagename,"wb") as f:
            f.write(response.content)
        i+=1
```
